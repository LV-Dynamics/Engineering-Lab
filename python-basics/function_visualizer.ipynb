# **Function Visualizer**

In this first section you can type a function and see it in a 2D space

### *How to write a function:*
- Write only the function expression (if the function is "$y=x+2$" write only "$x+2$")
- Little tips to write the expression:

| Math Symbol | How to write it |
|---|---|
| x + y | `x + y` |
| x - y | `x - y` |
| x × y | `x * y` |
| x ÷ y | `x / y` |
| x² | `x**2` |
| xⁿ | `x**n` |
| √x | `sqrt(x)` |
| !x! (absolute value) | `abs(x)` |
| sin(x) | `sin(x)` |
| cos(x) | `cos(x)` |
| tan(x) | `tan(x)` |
| e^x | `exp(x)` |
| ln(x) | `log(x)` |
| log10(x) | `log10(x)` |
| π | `pi` |
| e | `e` |

```python
import random
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

flist=[]

for f in range (0,int(input("How many function do you want to rapresent?"))):
    
    function = input('Write the function f(x)=...')
    print('f(x)='+ function)
    
    fy = sp.sympify("y-("+function+")")
    flist.append(fy)
    
    a = np.linspace(-10,10,100)
    b = eval(function, {"x":a, "sin":np.sin, "cos":np.cos, "sqrt":np.sqrt, "abs":abs, "pi":np.pi, "tan":np.tan, "ln":np.log, "log10":np.log10, "exp":np.exp, "e":np.e})
    
    color = (random.random(), random.random(), random.random())
    plt.plot(a,b,color=color)

x, y = sp.symbols('x y')
sol = sp.solve(flist, (x, y))
print("All solution "+str(sol))
    
plt.xlabel('X')
plt.ylabel('Y')
plt.ylim(-10,10)
plt.grid(True)
plt.show()
```
